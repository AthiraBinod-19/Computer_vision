```python
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread("IMG_2885.jpg")
```

```python
#Convert from BGR (OpenCV default) to RGB (Matplotlib expects RGB)
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```

```python
# Plot the image
plt.imshow(im_rgb)
plt.axis('off')       # hides axis for cleaner display
plt.show()
```

```python
#detecting the edges of this image using OPENCV
E = cv2.Canny(cv2.cvtColor(im,cv2.COLOR_BGR2GRAY),50,200)
```

```python
plt.imshow(E,cmap ='gray')
```
