**Edge Detection:**
It is done to simplify the visual data in an image while preserving the important structural information. by identifying and highlighting the edges, a CV system can reduce data volume and thereby makeing subsequent processing tasks more computationally effective and extract features and simplify analysis. Edge detection serves as foundational step for many advanced CV applications.
They are:
1) **Object recognition and detection:** Identifying the clear outlines of
objects(eg: cars, faces) is crucial for recognizing what they are.
2)**Image segmentation**: Dividing an image into meaningful, distinct regions(like seperating the foreground object from background)
3)**Medical imaging**: Highlighting boundaries of organs, tumors, etc. in scans to aid in diagnosis and analysis.
4)**Autonomous vehicles**: Detecting lane markings, road boundaries, and the outlines of other vehicles and pedestrians for navigation and safety.
5)**Robotics:** enabling robots to locate and grasp objects by precisely identifying their contours
6)**Feature extraction:** providing geometric features for tasks like face recognition and matching.

```python

im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
plt.axis('off')
plt.show()
```

```python
#detecting the edges of this image using OPENCV
E = cv2.Canny(cv2.cvtColor(im,cv2.COLOR_BGR2GRAY),50,200)
```

```python
plt.imshow(E,cmap ='gray')
```
